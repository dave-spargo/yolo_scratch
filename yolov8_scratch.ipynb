```python
import ultralytics
from ultralytics import YOLO

# Check version
print(f"YOLOv8 version: {ultralytics.__version__}")
```

```python
# This will download and organize coco128 dataset (for quick tests)
from ultralytics.utils.downloads import download
download('https://ultralytics.com/assets/coco128.zip', unzip=True)
```

```python
# Train a small model (YOLOv8n) for quick testing
model = YOLO("yolov8n.pt")  # other options: yolov8s.pt, yolov8m.pt, etc.

# Train on coco128
model.train(data="coco128.yaml", epochs=3, imgsz=640)
```

```python
results = model.val()
print(results)


```

```python
from PIL import Image

img_path = "/home/dave/Documents/general_AI/spargo_submit/coco128/images/train2017/000000000089.jpg"
img = Image.open(img_path)

# Run inference
results = model(img)

# Convert NumPy array to PIL Image before displaying
img_with_boxes = Image.fromarray(results[0].plot())

img_with_boxes.show()  # Opens in default image viewer
```
